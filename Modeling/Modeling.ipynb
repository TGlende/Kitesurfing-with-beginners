### Target

```python
X = df.drop('Kitesurf', axis = 1) 
y = df['Kitesurf']

# Train/Test sets
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 0, test_size = 0.4)

y_train.value_counts()
```

It seems we have a balanced sample

### Precision

We dim most important to eliminate FP (sending the newbies off to the beach with unfit weather conditions). With that in mind, we will choose the algorithm that gives us a better precision performance

```python
# Classifiers
classifiers = [SVC(), KNeighborsClassifier(), RandomForestClassifier(), GaussianNB()]

# Imputer: we can't afford to drop any rows 
knn = KNNImputer()

# Standard Scaler: our data have divers scaling
scaler = StandardScaler()

# Precision Performance
performance = pd.DataFrame( columns = ['SVM', 'KNN', 'RF', 'NB'], index = ['precision'])
precision = []

# Pipepline
for classifier in classifiers:
    pipe = Pipeline(steps=[('scaler', scaler), ('impute', knn), ('classifier', classifier)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    x = metrics.precision_score(y_test, y_pred)
    precision.append(x)

performance.loc['precision'] = precision
performance
```

Best performance is given by the random forest classifier. Let's check it out

### Random Forest

```python
# The model
model = Pipeline(steps=[('scaler', scaler), ('impute', knn), ('classifier', RandomForestClassifier(random_state = 2))])
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Score
model.score(X_test, y_test)
```

### Hyperparameter Tuning

We will use the GridSearchCV HT method as we want to be exhaustive in the hyperparameter search of our model

```python
max_features_s = [20, 30, 52]
criterion = ['gini', 'entropy']
max_depth = [3, 10, 20]
min_samples_split = [2, 4, 10]
min_samples_leaf = [1, 3]
max_features = ['auto', 'sqrt']

# Parameters Grid
param_grid = {'feature_selection__max_features' : max_features_s, 'classifier__criterion': criterion, 'classifier__max_depth': max_depth, 'classifier__min_samples_split': min_samples_split, 'classifier__min_samples_leaf': min_samples_leaf, 'classifier__max_features': max_features}


model = Pipeline(steps=[('scaler', scaler), ('impute', knn), ('feature_selection', SelectFromModel(RandomForestClassifier())), ('classifier', RandomForestClassifier())])

# Grid & fit
search = GridSearchCV(model, param_grid, scoring = 'precision')
search = search.fit(X_train, y_train)

# Results
search.best_params_
```

```python
# RandomForest
rf = RandomForestClassifier(criterion = 'entropy', max_depth = 3, max_features = 'sqrt', min_samples_leaf = 3, min_samples_split = 4, random_state = 0)

# The model
model = Pipeline(steps=[('scaler', scaler), ('impute', knn), ('feature_selection', SelectFromModel(rf, max_features = 20)), ('classifier', rf)])
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Precision
metrics.precision_score(y_test, y_pred)
```

We have better numbers but we still seek to better them. It is a question of safety after all! 

### Dimensionality Reduction

### Boosting

```python
# Classifiers
classifiers = [AdaBoostClassifier(), GradientBoostingClassifier(), XGBClassifier()]

# Imputer
knn = KNNImputer()

# Standard Scaler: our data have divers scaling
scaler = StandardScaler()

# Precision Performance
performance = pd.DataFrame( index = ['ADA', 'GB', 'XGB'], columns = ['precision'])
precision = []

# Pipepline
for classifier in classifiers:
    pipe = Pipeline(steps=[('scaler', scaler), ('impute', knn), ('classifier', classifier)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    x = metrics.precision_score(y_test, y_pred)
    precision.append(x)
    

performance['precision'] = precision
performance
```
